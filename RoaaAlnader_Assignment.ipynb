Calculate the Mean, Median and Mode of monthly last mile shipments, and explain what each metric means.

```python
monthly_mean = data.groupby('Month')['DaysUntilDelivery'].mean().reset_index()
monthly_mean.columns=["Month","Mean Value"]

monthly_median = data.groupby('Month')['DaysUntilDelivery'].median().reset_index()
monthly_median.columns=["Month","Median Value"]

monthly_mode = data.groupby('Month')['DaysUntilDelivery'].agg(mode).reset_index()
monthly_mode.columns=["Month","Mode Value"]
```

 1. **Mean:**
   - is the average and calculated by summing all the values in a dataset and dividing the sum by the total number of values.

2. **Median:**
   - is the middle value of a dataset when it is arranged in numerical order. If the number of values is even number. Then, the median is the average of the two middle values.

3. **Mode:**
   - is the value that appears most frequently in a dataset.

What months are below the mean and what months are above the mean and tell us what that could indicate?

```python
#1) calculate the mean for DaysUntilDelivery column
DaysUntilDelivery_mean = float(data['DaysUntilDelivery'].mean())
above_mean=[]
below_mean=[]

#2) Using the previous calculation -> the mean for each month -> monthly_mean
#3) checking every month if it above or under the DaysUntilDelivery mean
for i, row in monthly_mean.iterrows():
    if row["Mean Value"] > DaysUntilDelivery_mean:
        above_mean.append(row["Month"])
    if row["Mean Value"] < DaysUntilDelivery_mean:
        below_mean.append(row["Month"])
```

```python
#That tell us what the months that above the maen -> the months that has late delivery
above_mean
```

```python
#That tell us what the months that under the maen -> the months that has faster delivery
below_mean
```

```python
#1) Using the previous calculation -> the mean for each month -> monthly_mean
#2) Sort them -> first month is the highest and last month is the lowest
monthly_mean= monthly_mean.sort_values(by=['Mean Value'], ascending=False)
print("Highest:",monthly_mean["Month"].iloc[0])
print("Lowest:",monthly_mean["Month"].iloc[-1])
```

Can you define the interquartile range and calculate it for days until delivery?

```python
Q1= data["DaysUntilDelivery"].quantile(0.25)
Q2= data["DaysUntilDelivery"].quantile(0.50)
Q3= data["DaysUntilDelivery"].quantile(0.75)
IQR=Q3-Q1
print("IQR=",IQR)
```

**The interquartile range:** is the range of values between the first quartile (25th percentile) and the third quartile (75th percentile) in a dataset.






Can you specify outlier deliveries in days until delivery and what method you will use?

As in the following image.**Outliers** are the values that are larger than the maximum or smaller than the minimum




<img src="https://miro.medium.com/v2/resize:fit:4800/format:webp/1*wz--poSokEehPaqt6Kyxmw.png" width="600" height="300">


```python
maximum = (Q3+1.5*IQR)
minimum= (Q1-1.5*IQR)
print("maximum",maximum)
print("minimum", minimum)
outliers =[]
for i in data["DaysUntilDelivery"]:
    if (i > maximum or i < minimum):
        outliers.append(i)

print("----------------------------")
print("outliers:",outliers)
print("----------------------------")
```
